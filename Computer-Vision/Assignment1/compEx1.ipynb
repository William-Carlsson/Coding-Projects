```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
from mpl_toolkits.mplot3d import Axes3D
```

```python
def pflat(x):
    """Normalizes the input matrix by dividing by the last element of each column."""
    return x / x[-1]
```

```python
# Test pflat
x = np.array([[9,7],[3,-5],[3,-1]])

print(x.shape)
print(x)
ans = pflat(x)
print(ans.shape)
print(ans)
```

```python
# Load the data
data = scipy.io.loadmat('A1data/compEx1.mat')
print(data.keys())
#print(data["x2D"])
#print(data["x3D"])
print(type((data["x2D"])))
print(type((data["x3D"])))
print(data["x2D"].shape)
```

```python
pf_x2D = pflat(data["x2D"])
plt.figure(figsize=(10,8))
plt.scatter(pf_x2D[0,:], pf_x2D[1,:])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Scatter Plot of x2D Data")
plt.savefig("2Dplot")
plt.show()
```

```python
pf_x3D = pflat(data["x3D"])
fig =plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')
ax.scatter(pf_x3D[0, :], pf_x3D[1, :], pf_x3D[2, :], color='b', marker='o')
ax.set_xlabel("X-axis", fontweight ='bold')
ax.set_ylabel("Y-axis", fontweight ='bold')
ax.set_zlabel("Z-axis", fontweight ='bold')
plt.title("3D Scatter Plot of x3D Data")
fig.savefig("3Dplot")
plt.show()
```
