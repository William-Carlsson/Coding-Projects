```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from random import sample
```

```python
im1 = cv2.imread('data/cube1.jpg')
im2 = cv2.imread('data/cube2.jpg')
gray1 = cv2.cvtColor(im1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
```

```python
sift = cv2.SIFT_create()

kp1, des1 = sift.detectAndCompute(gray1, None)
kp2, des2 = sift.detectAndCompute(gray2, None)
```

```python
im1_with_kp = cv2.drawKeypoints(im1, kp1, None)
im2_with_kp = cv2.drawKeypoints(im2, kp2, None)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(im1_with_kp, cv2.COLOR_BGR2RGB))
plt.title("SIFT Features for cube1")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(im2_with_kp, cv2.COLOR_BGR2RGB))
plt.title("SIFT Features for cube2")

plt.show()
```

```python
bf = cv2.BFMatcher(cv2.NORM_L2, crossCheck=True)

matches = bf.match(des1, des2)
matches = sorted(matches, key = lambda x: x.distance)
```

```python
x1 = np.array([kp1[m.queryIdx].pt for m in matches])
x2 = np.array([kp2[m.trainIdx].pt for m in matches])
np.savez("x1_x2.npz", x1=x1, x2=x2)
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(np.hstack((im1, im2)))

for i in range(10):
    plt.plot([x1[i, 0], x2[i, 0] + im1.shape[1]], [x1[i, 1], x2[i, 1]], 'r-')

plt.title("Matching Points Between cube1 and cube2")
plt.show()
```
